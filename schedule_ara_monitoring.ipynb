```python
import pandas as pd
import numpy as np
import math

start = pd.Series(pd.date_range(start ='9/8/2020', end ='9/8/2022', freq='28D'))
end = pd.Series(pd.date_range(start ='10/5/2020', end ='10/5/2022', freq='28D'))
df = pd.DataFrame(dict(start_day = start,end_day = end ))

newdf = pd.DataFrame(np.repeat(df.values,2,axis=0))
newdf.columns = df.columns
Institute = ['Brussels','UCL', 'Wisconsin', 'VUB', 'UNL', 'Delaware', 'Maryland',
             'OSU', 'KU', 'Penn State', 'Chiba', 'NTU', 'Chicago',]

Station_num = [ 'ARA03-04', 'ARA01-02']
df1 = pd.DataFrame(Station_num) 
df2 = pd.DataFrame(Institute) 

newdf1 = pd.concat([df1]*int(len(newdf)/len(df1)), ignore_index=True) 
newdf2 = pd.concat([df2]*math.ceil(len(newdf)/len(df2)), ignore_index=True) 
newdf['station_number'] = newdf1 
newdf['institute'] = newdf2 
newdf
```

```python
iter = 0
for ind in newdf.index:
    if (newdf['institute'][ind] == 'Chicago'):
        iter += 1
        if (iter%3 == 1):
            newdf['institute'][ind] = 'KU'
        if (iter%3 == 2):
            newdf['institute'][ind] = 'OSU'
        if (iter%3 == 0):
            newdf['institute'][ind] = 'UW'  
```

```python
def mapper(month):
   return month.strftime("%d %b, %Y") 

newdf['start_day'] = newdf['start_day'].apply(mapper)
newdf['end_day'] = newdf['end_day'].apply(mapper)
```
